1. You have access to the salary information of several employees along with their Years of Experience. Using Linear regression analysis in machine learning, creating a linear regression model can predict the salary of an employee based on the years of experience.

# Importing Libraries To Build The Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import *
```

```python
x = df1.drop('Salary',axis=1)
y = df1['Salary']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25,random_state=42)
```

# Model Building

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
predicted_values = model.predict(x_test)
```

```python
result = pd.DataFrame({'Actual':y_test,'Predicted':predicted_values})
```

```python
result
```

```python
r2_score(y_test, predicted_values)
```

# Plotting Regression Plot

```python
sns.regplot(x = predicted_values, y = y_test)
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.title('Regression Plot')
plt.show()
```
